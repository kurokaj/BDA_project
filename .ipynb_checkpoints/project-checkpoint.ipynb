### Description

The Wisconsin breast cancer dataset contains diagnostic information from 569 measured instances. Each instance represents a digitized image of breast tissue extracted using fine needle aspiration (FNA). The features of each instance picture the distinctive traits of the nuclei in each image. In addition to the mean measurement value, this dataset also includes the standard error of the measurement and the mean of the three largest values (as the "worst" measurement). For this project only the mean values were selected in order to make the classification and analysis more straightforward. The attribute information is the following:

1. Diagnosis (M for malignant, B for benign)
2. Radius 
    - Mean of distances from center to points on perimeter
3. Texture 
    - Standard deviation of gray-scale values
4. Perimeter
5. Area
6. Smoothness 
    - Local variation in radius lengths
7. Compactness 
    - $\frac{Perimeter^2}{area} - 1.0$
8. Concavity
    - Severity of concave portions of contour
9. Concave points 
    - Number of concave portions of contour
10. Symmetry
11. Fractal dimension
    - $Coastline \approx -1$

The data set is relatively clean as it is. There is no missing values and zero values are rare and only in features 9 and 10. The binary class distribution is benign: 357 and malignant: 212. 

This dataset is relatively used for research and benchmarking purposes. We were able to locate multiple notebooks for example in Kaggle [1], that used this data set and applied a supervised learning model or clustering algorithm to it. Most of these approaches utilized some sort of graph based clustering or decision trees to assign the class labels. Our approach differs from them because we apply Bayesian workflow and propose a probabilistic model to calssify the data. One of the goals is to receive similar precision on the Bayesian model as were reveived for example on random forest classifier (~97%) [2]. In this project we also try to use set of different prior choises that have not been discussed in the literature.

```python
def train_test_split(data, labels, test_size=0.2, random_state=0):
    X_train, X_test, y_train, y_test = model_selection.train_test_split(data, 
                                                                        labels,
                                                                        test_size=test_size, 
                                                                        random_state=random_state)
    return X_train, X_test, y_train, y_test
```

## Model

In this work multiple logistic models with different prior choices are compared in above mentioned binary classification task. All of the models assume linear correlation between features and the predicted value:

$$\mathbf{y} = x_1\beta_1 + x_2\beta_2 + ... + x_n\beta_n + \epsilon = \mathbf{X\beta} + \epsilon$$

The residuals $\epsilon$ are assumed similarly distributed as the $\beta$- coefficients. The likelihoods of each models are modelled using single success trial bernoulli- distribution and can be written as: 

$$\mathbf{y}|\mathbf{X,\beta} \sim Bernoulli(\mathbf{X\beta})$$

By the Bayesian justification the posterior distribution for each of the models is the product of likelihood and prior distribution:

$$p(\mathbf{\beta}|\mathbf{y,X}) \propto p(\mathbf{y}|\mathbf{X,\beta}) \times p(\beta), $$

where the prior is different for each model. The Stan- scripts of each models are presented below. They all share the same data block and generative block.

### Priors

The prior choice in Bayesian modelling plays large role. It defines how much of the load is placed on the information of the data that is seen before the modelling. Generally it is wise to avoid using flat or super uninformative priors as that approach often puts too much weight on only the modelled likelihood. Weakly informative priors contain some information, at least enough to regularize the posterior distribution. [1][2] 

In this work is presented total three different prior choices: nonconjugate normal, t- distribution and cauchy- distribution prior. For the nonconjugate prior model the prior distribution is chosen to be generic weakly informative prior $\beta_i \sim normal(0, 1)$. Also the residual is drawn from similar distribution. This way the prior utilizes a little information from the data, in order to slightly constrain the posterior. Using weakly informative prior also simplifyes the computations.  

In the t-distribution prior models the prior is similarly to first one a weakly informative prior $\beta_i \sim normal(0, std)$. However the standard deviation is unknown and is drawn from inverse chi-squared -distribution with 2 degrees of freedom. T- distribution is like normal distribution, but it has heavier tales, which means that it is more prone to give values that lie further from its mean. With this apporach we will be able to get larger scale of standard deviation values than in the first model. Out of these two prior choises the T- prior is more informative. 

The third prior choice for this project is cauchy- distributed prior. It is very similar to normal distribution, but has wider tales similarly to t-distribution. With these models the coefficients are drawn $\beta_i \sim cauchy(0, 2)$.

Using wider and heavy- taled priors, we make the models more robust and usable in the CVLOO comparison. Linear regression is normally sensitive to outliers as the normal distribution is rather narrow by default. Using wider priors with wider tails decreases the effect of outliers in the log- posterior of the model. [3] Different values for t- distribution and cauchys distribution were used and wider priors were noted to produce a little better models.  

### Logistic regression

Logistic regression is in many ways just like linear model. The features are assumed to be linearly correlated and out comes binary. The link function between probabilities and and log odds is: 
$$logit(v) = log(\frac{v}{1-v})$$, 
and its inverse:
$$logit^{-1}(u) = log(\frac{1}{1+exp(-u)})$$
The models are formulated in Pystan and use the logit- parametrized Bernoulli distributions:
$$BernoulliLogit(y|\beta,\alpha) = Bernoulli(y|logit^{-1}(\beta,\alpha))$$

### Stan code of the models

```python
models = dict()
```

#### Common blocks

```python
data_block =    """
                data {
                  int<lower=0> N;
                  int<lower=0> T;
                  int<lower=0> J;

                  matrix[N, J] x;
                  matrix[T, J] t;

                  int<lower=0, upper=1> y[N]; 
                }
                """

generated_block =   """
                    generated quantities {
                      vector[N] log_lik;
                      int test_pred[T];
                      // int y_hat[N];
                      
                      for (i in 1:N) {
                         log_lik[i] = bernoulli_logit_lpmf(y[i] | alpha + x[i,] * beta);
                      }
                      test_pred = bernoulli_logit_rng(alpha + t * beta);
                      // y_hat = bernoulli_logit_rng(alpha + x * beta);
                    }
                    """
```

#### Normal prior

```python
normal_code = data_block + """
                        parameters {
                          real alpha;
                          vector[J] beta;
                        }
                        model {
                          alpha ~ normal(0, 1);
                          beta ~ normal(0, 1);
                          
                          y ~ bernoulli_logit(alpha + x * beta);
                        }
                        """ + generated_block

normal_model = pystan.StanModel(model_code=normal_code)
models["normal"] = normal_model
```

#### t prior

```python
t_prior_code =  data_block + """
                parameters {
                  real alpha;
                  vector[J] beta;
                  
                  real<lower=0> sigma_a;
                  vector<lower=0>[J] sigma_b;
                }
                model {
                  sigma_a ~ inv_chi_square(2);
                  for (j in 1:J) {
                    sigma_b[j] ~ inv_chi_square(2);
                  }

                  alpha ~ normal(0, sigma_a);
                  for (j in 1:J){
                    beta[j] ~ normal(0, sigma_b[j]);
                  }

                  y ~ bernoulli_logit(alpha + x * beta);
                }
                """ + generated_block

t_prior_model = pystan.StanModel(model_code=t_prior_code)
models["t"]= t_prior_model
```

#### Cauchy prior

```python
cauchy_prior_code =  data_block + """
                parameters {
                  real alpha;
                  vector[J] beta;
                }
                model {
                  alpha ~ cauchy(0, 2);
                  beta ~ cauchy(0, 2);
                  y ~ bernoulli_logit(alpha + x * beta);
                }
                """ + generated_block

cauchy_model = pystan.StanModel(model_code=cauchy_prior_code)
models["cauchy"]= cauchy_model
```

## Experimenting

### Sampling

```python
def sample_model(model, data, iter_count=2000):
    return model.sampling(data=data, iter=iter_count, chains=4, control={'max_treedepth': 15, 'adapt_delta' : 0.9}) #control={"max_treedepth":10}


def run_experiment(datasets, models, iter_count=2000):
    fits = dict()
    current = 1
    total = len(datasets.keys()) * len(models.keys())
    for dataset in datasets.keys():
        fits[dataset] = dict()
        for model in models.keys():
            print("Sampling {}/{}: Data = {}, Model = {}".format(current, total, dataset, model))
            fits[dataset][model] = sample_model(models[model],
                                                datasets[dataset].stan_dict,
                                                iter_count)
            current += 1
    return fits

fits = run_experiment(datasets, models)
```

### Posterior predictive check

The posterior predictive check (PPC) was performed to every model and data set choices and they are plotted in appendix 1. The purpose of PPC is to simulate the datasets from the posterior predictive distribution. In other words the model generates new instances of data based on the model that is built until that moment. From the draws of posterior predictive distribution it is possible to evaluate the if the model structure is plausible. The predicted posterior is formulated: 

$$p(\tilde{y}|y) = \int p(\tilde{y}|\theta) \times p(\theta|y) d\theta$$

The PPC is marginalizing over posterior inferences in order to get the predicted $\tilde{y}$ given the observed $y$ from the dataset. In the appendix 1 is plotted the mean data overlaid by posterior predictive draws. As can be seen the PPC mean is at the same or at least near the observed y in every model. This means that the models are indeed plausible. This could have been noticed also from the test dataset predictive distribution and that the models reach nice predictability levels.  

```python
# Plot all posterior predictive checks to one large plot -> add to appendices
[fig, axes] = plt.subplots(3, 5)

for m,model in enumerate(models.keys()):
    for d,dataset in enumerate(datasets.keys()): 
        print(model + " --- " + dataset)
        fit_ppc = fits[dataset][model] 
        data_ppc = az.from_pystan(
                posterior=fit_ppc,
                posterior_predictive='y_hat',
                observed_data=['y'],
             )
        az.plot_ppc(data_ppc, data_pairs = {'y' : 'y_hat'}, ax= axes[m,d])
        axes[m,d].set_title(("PPC for model: "+ model + " with data: " + dataset))

fig.set_figheight(20)
fig.set_figwidth(35)
fig.savefig('PPC_all.png')
```

### Evaluation (A little bit annoying...)

```python
def get_idata(fit):
    return az.from_pystan(fit, log_likelihood="log_lik")

def get_idata_dict(fits_dict):
    idata_dict = dict()
    for dataset in fits_dict.keys():
        for model in fits_dict[dataset].keys():
            key = "{}-{}".format(dataset, model)            
            idata_dict[key] = get_idata(fits_dict[dataset][model])
    return idata_dict

def get_loo(fit):
    return az.loo(get_idata(fit), pointwise=True, var_name= "log_lik")

def get_test_samples(fit):
    return fit.extract(pars="test_pred")["test_pred"]

def get_ytest_ypred(fit, y_test):
    y_pred = get_test_samples(fit)
    y_test_copied = np.dstack([y_test]*y_pred.shape[0])[0].T
    return np.array(y_pred, dtype=int).flatten(), np.array(y_test_copied, dtype=int).flatten()

def get_pred_accuracy(fit, y_test):
    y_test, y_pred = get_ytest_ypred(fit, y_test)
    return metrics.accuracy_score(y_test, y_pred)

def get_pred_accuracy_all(fits_dict, datasets):
    accuracy_dict = dict()
    for dataset in fits_dict.keys():
        for model in fits_dict[dataset].keys():
            key = "{}-{}".format(dataset, model)
            accuracy_dict[key] = get_pred_accuracy(fits_dict[dataset][model], 
                                                   datasets[dataset].y_test)
    return accuracy_dict

def compare_models(fits_dict, datasets):
    compare_frame = az.compare(get_idata_dict(fits_dict))
    accuracy_dict = get_pred_accuracy_all(fits_dict, datasets)
    compare_frame["pred_accuracy"] = pd.Series(accuracy_dict)
    return compare_frame

compare_frame = compare_models(fits, datasets)
display(compare_frame)
```

### Best models details

```python
SEPARATOR = "\n=================================================================================================\n"
def print_loo_stats(fit):
    print(get_loo(fit))

def print_summary_stats(fit):
    display(az.summary(fit, var_names=["alpha", "beta"], fmt="wide", kind="stats"))
    
def print_summary_diagnostics(fit):
    display(az.summary(fit, var_names=["alpha", "beta"], fmt="wide", kind="diagnostics"))

def plot_khat(fit):
    az.plot_khat(get_loo(fit))
    plt.show()
    
def print_classification_report(fit, y_test):
    y_test, y_pred = get_ytest_ypred(fit, y_test)
    
    print(metrics.classification_report(y_test, y_pred))
    metrics.ConfusionMatrixDisplay(metrics.confusion_matrix(y_test, y_pred)).plot()
    plt.show()
    

def print_all_details(fit, dataset):
    print_summary_stats(fit)
    print(SEPARATOR)
    print_loo_stats(fit)      
    plot_khat(fit)
    print(SEPARATOR)
    print_summary_diagnostics(fit)
    print(SEPARATOR)
    print_classification_report(fit, dataset.y_test)
    
    
best_model = fits["All st"]["t"]

print_all_details(best_model, datasets["All st"])
```
