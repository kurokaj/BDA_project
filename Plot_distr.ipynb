```python
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np
from scipy.stats import t
from scipy.stats import cauchy


x_min = -5
x_max = 5

mu = 0 
gamma = 2

df = 2

mean = 0
std = 1

x = np.linspace(x_min, x_max, 100)
dist = cauchy(mu, gamma)
dist2 = t(2)

y = scipy.stats.norm.pdf(x,mean,std)
y2 = dist.pdf(x)
y3 = dist2.pdf(x)

plt.plot(x,y, color='coral', label="Normal(0,1)- prior")
plt.plot(x,y2, color='black', label="Cauchy(2)- prior")
plt.plot(x,y3, color='lime', label="Students t(2)- prior")


plt.legend()
plt.grid()
```
